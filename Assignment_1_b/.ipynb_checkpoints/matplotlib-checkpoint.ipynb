```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
plt.plot([1,2,3,4]) #single array will be treated as y axis data.
plt.ylabel('numbers')
plt.show()
#x axis will start from 0.
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
# x verses y
```

```python
#style of plot and use of axis
plt.plot([1,2,3,4],[1,4,9,16],'ro')  # 'ro' means red dot and the default blue line is 'b-'
plt.axis([0,6,0,20])
plt.show()
# xy.axis([xmin,xmax,ymin,ymax])
```

```python
#numpy in matplotlib
t=np.arange(0.,5.,0.2)
plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^')
plt.show()
#'r--' is red dash, 'bs' means blue squares, 'g^' means green triangles
```

```python
#plotting with keyword strings
data={'a':np.arange(50),
     'c':np.random.randint(0,50,50),
     'd':np.random.randn(50)}
data['b']=data['a']+10*np.random.randn(50)
data['d']=np.abs(data['d'])*100
data
```

```python
plt.scatter('a','b',c='c',s='d',data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
#don't know what is .scatter()
```

```python
#plotting with categorical variables
names=['group a','group b','group c']
values=[1,10,100]
plt.figure(figsize=(9,3))
plt.subplot(131)
plt.bar(names,values)
plt.subplot(132)
plt.scatter(names,values)
plt.subplot(133)
plt.plot(names,values)
plt.suptitle('Categorical Plotting')
plt.show()
```

```python
#plotting with categorical variables
names=['group a','group b','group c']
values=[1,10,100]
plt.figure(figsize=(9,3))
plt.subplot(131)
plt.bar(names,values)
plt.subplot(131)
plt.scatter(names,values)
plt.subplot(131)
plt.plot(names,values)
plt.suptitle('Categorical Plotting')
plt.show()
#all on same graph..notice the argument in .subplot()
```

```python
#multiple figures and axes
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
t1=np.arange(0.,5.,0.1)
t2=np.arange(0.,5.,0.02)
plt.figure()
plt.subplot(211)
plt.plot(t1,f(t1),'bo',t2,f(t2),'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2),'r--')
plt.show()
```
